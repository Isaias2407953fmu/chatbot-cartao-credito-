**Esta versão implementa leitura do arquivo de perguntas.csv**

```python
#bibliotecas usadas no projeto
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd

#CountVectorizer é uma classe da biblioteca sklearn que transforma
#textos do vocabulário em vetores numéricos.
#MultinomialNB classifica os textos com base na frequência das palavras.
#Aprende com a contagem das palavras vetorizadas.
#Faz predições com base nas probabilidades aprendidas.

#DADOS DE TREINAMENTO

perguntas = pd.read_csv("/content/perguntas.csv")

frases = perguntas["frase"].astype(str).tolist()
categorias = perguntas["categoria"].astype(str).tolist()


#VETORIZAÇÃO E TREINAMENTO
vetorizador = CountVectorizer()
X = vetorizador.fit_transform(frases)

#vetorizador vai guardar um objeto que será usado para transformar texto em número.
#X guarda o resultado da transformação das frases em vetores numéricos.
#fit_transform é um método da classe CountVectorizer, ele aprende o vocabulário
#das frases e transforma em vetor.

modelo = MultinomialNB()
modelo.fit(X, categorias)
#modelo vai guardar o modelo criado
#modelo.fit - algoritmo que treina com os exemplos das frases que passamos pra ele

#RESPOSTAS
respostas = {
    'limite': 'Você pode consultar o limite disponível no aplicativo ou na área do cliente.',
    'fatura': 'Sua fatura pode ser consultada no aplicativo, no site ou na central de atendimento.',
    'bloqueio': 'Por segurança, bloqueie o cartão imediatamente pelo aplicativo ou entre em contato com a central.',
    'desbloqueio': 'O desbloqueio do cartão pode ser feito no aplicativo após a confirmação dos seus dados.',
    'parcelamento': 'O parcelamento da fatura pode ser solicitado pelo aplicativo ou com a central de atendimento.',
    'vencimento': 'A data de vencimento da fatura pode ser consultada no aplicativo, no boleto ou na área do cliente.',
    'cancelamento': 'Entre no aplicativo e faça o cancelamento do seu cartão com segurança',
    'senha': 'Já enviei para seu email um link para criar uma nova senha'
    }

#CHATBOT
print('='*45)
print('  CHATBOT OPERADORA DE CARTÃO DE CRÉDITO   ')
print("Digite sua pergunta ou 'sair' para encerrar")
print('='*45)

while True:
  pergunta = input('\nVocê: ').lower()

  if pergunta == 'sair':
    print('Chatbot: Atendimento encerrado.')
    break #sai do laço

  pergunta_vetorizada = vetorizador.transform([pergunta])
  #pergunta_vetorizada transforma a pergunta digitada em uma forma
  #numérica que o modelo consegue entender.
  #vetorizador é o objeto CountVectorizer(), que foi treinado antes
  #com as frases.

  categoria_prevista = modelo.predict(pergunta_vetorizada)[0]
  #modelo.predict() analisa o vetor da pergunta e decide a qual
  #categoria pertence (começa do zero)

  probabilidades = modelo.predict_proba(pergunta_vetorizada)[0]
  #retorna a probabilidade de cada categoria (0.10, 0.75, 0.15) em porcentagem

  maior_probabilidade = max(probabilidades)
  #guarda a maior probabilidade de resposta escolhida pelo modelo

  if maior_probabilidade < 0.40: #menor que 40%
      print("Chatbot: Desculpe, não entendi sua solicitação. Pode reformular a pergunta?")
  else:
      print("Categoria identificada:", categoria_prevista)
      print("Probabilidade:", round(maior_probabilidade * 100, 2), "%")
      print("Chatbot:", respostas[categoria_prevista])
```
